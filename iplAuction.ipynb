```python
g1=data.groupby('player style')['base price (in lacs)'].mean().idxmax()
g2=data.groupby('player style')['base price (in lacs)'].mean().max()
g3=data.groupby('player style')['base price (in lacs)'].mean().idxmin()
g4=data.groupby('player style')['base price (in lacs)'].mean().min()



print(f'The playing style {g1} has the highest base price on avg of {round(g2,2)} lacs')
print(f'The playing style {g3} has the lowest base price on avg of {round(g4,2)} lacs')
```

```python
# ASSUMING DATA IS YOUR DATAFRAME COTAINING DATA ABOUT STATUS AND FRANCHISE
# GROUPING BT 'STATUS ' AND 'FRANCHISE ', AND COUNTING OCCURRENCES
st = data.groupby('status')['franchise'].value_counts()


# SELECTING COUNTS FOR RETAINED STATUS
st1=st.loc['RETAINED']

# FINDING THE MAX COUNT OF FRANCHISE FOR 'RETAINED' STATUS
max_count = st1.max()

print("Maximum count of franchises for 'RETAINED' status:" ,max_count)
```
